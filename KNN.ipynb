```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
col_names= ['sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'class']
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
dataset = pd.read_csv(url, names=col_names)
```

```python
dataset
```

```python
dataset_knn = dataset.iloc[:, 0:4]
labels = dataset.iloc[:, -1]
```

```python
dataset_knn
```

```python
import math
```

```python
import numpy as np

def distance(x1, x2):
    return np.linalg.norm(np.array(x1) - np.array(x2))
```

```python
def distance(x1,x2):
  distances = 0.0
  for i in range(len(x1)):
    distances += (x1[i] - x2[i]) ** 2
  return (math.sqrt(distances))
```

```python
testing_data = dataset.values[101, 0:4]
```

```python
distance_cumm = []
for i in range(len(dataset_knn)):
  distance_cumm.append([distance(dataset_knn[i], testing_data), labels[i]])
```

```python
distance_cumm = [
    [distance(dataset_knn[i], testing_data), labels[i]]
    for i in range(len(dataset_knn))
]
```

```python
distance_cumm = [
    [distance(dataset_knn.iloc[i], testing_data), labels.iloc[i]]
    for i in range(len(dataset_knn))
]
```

```python
distance_cumm = []
for row, label in zip(dataset_knn.itertuples(index=False), labels):
    distance_cumm.append([distance(row, testing_data), label])
```

```python
distance_cumm
```

```python
distance_cumm.sort()
```

```python
distance_cumm
```

```python
k=5
first_k=[]
for i in range(k):
  first_k.append(distance_cumm[i][1])
```

```python
first_k

```

```python
from collections import Counter
count = Counter(first_k)
```
