```python
#import bsic libaries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
  return (1/(1+np.exp(-x)))
```

```python
x=np.linspace(-10,10,200)
def plot_function(x,y,title):
  plt.plot(x,y,label=title)
  plt.title(title)
  plt.show()
```

```python
plot_function(x,sigmoid(x),"Sigmoid")
```

```python
#implementation of the step function
#implementation of ReLU fun
```

```python
def step(x):
  return np.where(x>=0,1,0)
```

```python
plot_function(x,step(x),"Step")
```

```python
def relu(x):
  return np.maximum(0,x)
```

```python
plot_function(x,relu(x),"Relu")
```

```python
plot_function(x,relu(x),"ReLU")
```

```python
def tanh(x):
  return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
plot_function(x,tanh(x),"TanH")
```

```python
def softmax(x):
  e1=(np.exp(x-np.max(x)))
  return e1/np.sum(e1,axis=0)
```

```python
z=np.array([0.51,0.48,0.01])
softmax(z)
```
